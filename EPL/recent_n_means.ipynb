```python
def recent_5_mean(data, group_var, variable, lag_num):
    final_data = pd.DataFrame()
    for var in variable:
       lag_data = data.groupby(group_var)[var].shift().rolling(lag_num).mean()
       final_data[var] = lag_data
    return final_data

sum_var = unov.columns[3:27].tolist()
group_var = ['Home']
group_var2 = ['Home','Date']
sum_var 
```
