# EPL

```python
# 원본 데이터
data = pd.read_excel('EPL.xlsx')
data.head()
```

```python
data.shape
```

```python
# Data 컬럼의 값을 betman 엑셀파일의 날짜 형식과 동일하게 변경
data['Date'] = pd.to_datetime(data['Date'], format='%a %d %b %Y').dt.strftime('%y.%m.%d (%a)')

# 요일 매핑 딕셔너리 생성
day_mapping = {
    'Mon': '월',
    'Tue': '화',
    'Wed': '수',
    'Thu': '목',
    'Fri': '금',
    'Sat': '토',
    'Sun': '일'
}

# 'Date' 컬럼 요일 한글로 변경
data['Date'] = data['Date'].str.replace(r'\((\w+)\)', lambda x: f'({day_mapping[x.group(1)]})')

data['team&Stats'] = data['team&Stats'].str[17:-2]

# Home, Away 컬럼 추가
data['Home'] = data['team&Stats'].str.split('\'').str[0]
data['Away'] = data['team&Stats'].str.split('\'').str[2]
data['team&Stats'] = data['team&Stats'].str.split("'", n=4).str[-1]

# 홈, 어웨이 점유율 컬럼 추가
data['H_Possession'] = data[data['team&Stats'].str.contains('Possession', na=False)]['team&Stats'].str.split(' ').str[0]
data['A_Possession'] = data[data['team&Stats'].str.contains('Possession', na=False)]['team&Stats'].str.split('% ').str[1].str.split("'").str[0]
data.loc[data['team&Stats'].str.contains('Possession', na=False), 'team&Stats'] = data['team&Stats'].str.split("'", n=2).str[-1]

# 홈, 어웨이 유효슈팅 수 컬럼 추가
data['H_Shots on target'] = data[data['team&Stats'].str.contains('Shots on target', na=False)]['team&Stats'].str.split(' ').str[0]
data['A_Shots on target'] = data[data['team&Stats'].str.contains('Shots on target', na=False)]['team&Stats'].str.split(' ', n=4).str[4].str.split("'").str[0]
data.loc[data['team&Stats'].str.contains('Shots on target', na=False), 'team&Stats'] = data['team&Stats'].str.split("'", n=2).str[-1]

# 홈, 어웨이 슈팅 수 컬럼 추가
data['H_Shots'] = data[data['team&Stats'].str.contains('Shots', na=False)]['team&Stats'].str.split(' ').str[0]
data['A_Shots'] = data[data['team&Stats'].str.contains('Shots', na=False)]['team&Stats'].str.split(' ', n=2).str[2].str.split("'").str[0]
data.loc[data['team&Stats'].str.contains('Shots', na=False), 'team&Stats'] = data['team&Stats'].str.split("'", n=2).str[-1]

# 홈, 어웨이 터치 수 컬럼 추가
data['H_Touches'] = data[data['team&Stats'].str.contains('Touches', na=False)]['team&Stats'].str.split(' ').str[0]
data['A_Touches'] = data[data['team&Stats'].str.contains('Touches', na=False)]['team&Stats'].str.split(' ', n=2).str[2].str.split("'").str[0]
data.loc[data['team&Stats'].str.contains('Touches', na=False), 'team&Stats'] = data['team&Stats'].str.split("'", n=2).str[-1]

# 홈, 어웨이 패스 수 컬럼 추가
data['H_Passes'] = data[data['team&Stats'].str.contains('Passes', na=False)]['team&Stats'].str.split(' ').str[0]
data['A_Passes'] = data[data['team&Stats'].str.contains('Passes', na=False)]['team&Stats'].str.split(' ', n=2).str[2].str.split("'").str[0]
data.loc[data['team&Stats'].str.contains('Passes', na=False), 'team&Stats'] = data['team&Stats'].str.split("'", n=2).str[-1]

# 홈, 어웨이 태클 수 컬럼 추가
data['H_Tackles'] = data[data['team&Stats'].str.contains('Tackles', na=False)]['team&Stats'].str.split(' ').str[0]
data['A_Tackles'] = data[data['team&Stats'].str.contains('Tackles', na=False)]['team&Stats'].str.split(' ', n=2).str[2].str.split("'").str[0]
data.loc[data['team&Stats'].str.contains('Tackles', na=False), 'team&Stats'] = data['team&Stats'].str.split("'", n=2).str[-1]

# 홈, 어웨이 걷어내기 수 컬럼 추가
data['H_Clearances'] = data[data['team&Stats'].str.contains('Clearances', na=False)]['team&Stats'].str.split(' ').str[0]
data['A_Clearances'] = data[data['team&Stats'].str.contains('Clearances', na=False)]['team&Stats'].str.split(' ', n=2).str[2].str.split("'").str[0]
data.loc[data['team&Stats'].str.contains('Clearances', na=False), 'team&Stats'] = data['team&Stats'].str.split("'", n=2).str[-1]

# 홈, 어웨이 코너킥 수 컬럼 추가
data['H_Corners'] = data[data['team&Stats'].str.contains('Corners', na=False)]['team&Stats'].str.split(' ').str[0]
data['A_Corners'] = data[data['team&Stats'].str.contains('Corners', na=False)]['team&Stats'].str.split(' ', n=2).str[2].str.split("'").str[0]
data.loc[data['team&Stats'].str.contains('Corners', na=False), 'team&Stats'] = data['team&Stats'].str.split("'", n=2).str[-1]

# 홈, 어웨이 오프사이드 수 컬럼 추가
data['H_Offsides'] = data[data['team&Stats'].str.contains('Offsides', na=False)]['team&Stats'].str.split(' ').str[0]
data['A_Offsides'] = data[data['team&Stats'].str.contains('Offsides', na=False)]['team&Stats'].str.split(' ', n=2).str[2].str.split("'").str[0]
data.loc[data['team&Stats'].str.contains('Offsides', na=False), 'team&Stats'] = data['team&Stats'].str.split("'", n=2).str[-1]

# 홈, 어웨이 경고 수 컬럼 추가
data['H_Yellow cards'] = data[data['team&Stats'].str.contains('Yellow cards', na=False)]['team&Stats'].str.split(' ').str[0]
data['A_Yellow cards'] = data[data['team&Stats'].str.contains('Yellow cards', na=False)]['team&Stats'].str.split(' ', n=3).str[3].str.split("'").str[0]
data.loc[data['team&Stats'].str.contains('Yellow cards', na=False), 'team&Stats'] = data['team&Stats'].str.split("'", n=2).str[-1]

# 홈, 어웨이 퇴장 수 컬럼 추가
data['H_Red cards'] = data[data['team&Stats'].str.contains('Red cards', na=False)]['team&Stats'].str.split(' ').str[0]
data['A_Red cards'] = data[data['team&Stats'].str.contains('Red cards', na=False)]['team&Stats'].str.split(' ', n=3).str[3].str.split("'").str[0]
data.loc[data['team&Stats'].str.contains('Red cards', na=False), 'team&Stats'] = data['team&Stats'].str.split("'", n=2).str[-1]

# 홈, 어웨이 파울 수 컬럼 추가
data['H_Fouls'] = data[data['team&Stats'].str.contains(' Fouls conceded ', na=False)]['team&Stats'].str.split(' ').str[0]
data['A_Fouls'] = data[data['team&Stats'].str.contains(' Fouls conceded ', na=False)]['team&Stats'].str.split(' ', n=3).str[3].str.split("'").str[0]

data.drop('team&Stats', axis=1, inplace=True)
data
```

# XG

```python
from datetime import datetime
```

```python
xg = pd.read_excel('XG.xlsx')
xg
```

```python
# 날짜 형식 변경
xg["data"]=list(map(lambda x: datetime.strptime(x,"%A, %B %d, %Y").strftime("%y.%m.%d (%A)"), xg['data']))
```

```python
# 나누고 새로운 컬럼 생성
new_columns = ['Team', 'Score', 'Ratio', 'Opponent']

# 각 행에 대해 새로운 열 생성
xg[new_columns] = xg['Game'].apply(lambda row: pd.Series(eval(row)))

# 결과 확인
xg.head()
```

```python
# 굳이 필요는 없음 10골 이상 안들어가서
def split_scores(score):
    if len(score) == 2:
        return score[0], score[1]
    else:
        return None, None

# "H_score"와 "A_score" 열 생성
xg[["H_score", "A_score"]] = xg["Score"].apply(lambda score: pd.Series(split_scores(score)))

# 결과 확인
xg.head()
```

```python
# "H_ratio"와 "A_ratio" 열 생성
xg[["H_ratio", "A_ratio"]] = xg["Ratio"].apply(lambda ratio: pd.Series([ratio[:4], ratio[4:]]))

# 결과 확인
xg.head()
```

```python
# 팀명 통합
team_mapping = {
'Aston Villa' : 'Aston Villa'
,'Manchester City' : 'Manchester City'
,'Norwich' : 'Norwich City'
,'Tottenham' : 'Tottenham Hotspur'
,'Watford' : 'Watford'
,'Brighton' : 'Brighton & Hove Albion'
,'West Ham' : 'West Ham United'
,'Bournemouth' : 'Bournemouth'
,'Newcastle United' : 'Newcastle United'
,'Everton' : 'Everton'
,'Leicester' : 'Leicester City'
,'Manchester United' : 'Manchester United'
,'Liverpool' : 'Liverpool'
,'Burnley' : 'Burnley'
,'Southampton' :  'Southampton'
,'Chelsea' : 'Chelsea'
,'Arsenal' : 'Arsenal'
,'Sheffield United' : 'Sheffield United'
,'Fulham' : 'Fulham'
,'West Bromwich Albion' : 'West Bromwich Albion'
,'Leeds' : 'Leeds United'
,'Wolverhampton Wanderers' : 'Wolverhampton Wanderers'
,'Brentford' : 'Brentford'
,'Crystal Palace' : 'Crystal Palace'
,'Nottingham Forest' : 'Nottingham Forest'
}

xg["Team"] = xg["Team"].map(team_mapping)
xg["Opponent"] = xg["Opponent"].map(team_mapping)
xg.head()
```

```python
xg.drop(["Game","Score","Ratio"],axis=1,inplace=True)
```
