```python
team = list(unov['Home'].unique())
```

### ex라는 데이터 프레임에 각 팀별로 롤링, 평균 구하고 합치기

```python
ex = pd.DataFrame()

for h in team:
    for col in cols:
        t = unov.groupby(by = 'Home').get_group(h)[col].rolling(3).mean().shift(-2)
        ex = pd.concat([ex,t])
        # for idx, val in enumerate(t):
        
        # ex.loc[t.index, col] = t
    # ex = pd.concat([ex,t])
    
```
